```python
import numpy as np
np.pi
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

theta_in_radians = np.radians(range(0,365,5))
theta_in_radians
```

```python
mytans = np.tan(theta_in_radians)
mytans
```

```python
plt.plot(theta_in_radians,mytans, "r", label="Tangent");
plt.ylim([-1,1])
plt.xlabel("Radians")
plt.ylabel("Tangent")
plt.title("The Transcedental Tan")
plt.legend(loc="center");
```

```python
import os
mynums = list(range(1931,2017,1))
print(mynums[2:5])
rainfall = np.loadtxt('../Datasets/rainFall/Avg_Rain.txt')
print (rainfall)
```

```python
plt.plot(mynums,rainfall, label = "Annual Rainfall, inches")
plt.xlabel('Year')
plt.ylabel("Annual Rainfall")
plt.legend()
```

```python
z = np.polyfit(mynums, rainfall, 1) 
p = np.poly1d(z)
year = mynums
plt.plot(year,p(year))
plt.plot(year, rainfall)
```
