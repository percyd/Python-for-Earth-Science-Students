```python
EQ=np.loadtxt('../Datasets/seismicRecord/earthquake.txt') # read in data
time_units=np.arange(len(EQ)) # makes an array of arbitrary time units same length as data
minutes = time_units/1200
plt.plot(minutes, EQ,'r-') # plots as a red line
plt.xlabel('Time (Minutes)') # puts a label on the X axis
plt.ylabel('Velocity'); # puts a label on the Y axis
plt.title('Earthquake observed at Pinyon Flat Observatory');
plt.text(4,10000,'P-S delay interval', rotation = 90, c='b', va='bottom')
```
