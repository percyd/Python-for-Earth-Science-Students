```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
np_load = True
X = np.array([[2, 1], [3, 2], [4, 2], [5, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3
lr = LinearRegression().fit(X, y)
myprobe = lr.predict(np.array([[1, 5]]))
```
